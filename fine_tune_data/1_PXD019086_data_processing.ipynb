```python
####  HeLa_trypsin
## 提取出evidence文件中的RT以及CCS数据，添加到msms文件中，然后删除不需要的数据
import pandas as pd
from tqdm import tqdm
data_file = r"PXD019086\Results_HeLa_trypsin"
## 读取evidence数据
evidence_path = data_file + '/evidence.txt'
with open(evidence_path,'r', encoding='utf8') as fa:
     edata = fa.read()
evidence = pd.DataFrame([line.split('	') for line in edata.split("\n")])   
## 读取msms数据
msms_path = data_file + '/msms.txt'
with open(msms_path,'r', encoding='utf8') as fa:
     mdata = fa.read()
msms = pd.DataFrame([line.split('	') for line in mdata.split("\n")])   
```

```python
# 删除最后一行
evidence.drop(index=evidence.index[-1], axis=0, inplace=True)
msms.drop(index=msms.index[-1], axis=0, inplace=True)
## 由于evidence文件与msms文件每行数据是一一对应的，可以直接合并
Data = pd.concat([msms, evidence],axis=1)
Data.columns = list(range(len(Data.columns)))
## 只保留需要的数据
list_d = [0,3,4,7,13,14,15,16,19,20,21,22,24,26,39,40,41,42,43,44,45,46,73,95,98,113,117]
Data = Data.iloc[:,list_d]
## 将第一行数据变成列名称，删除最后一行并重新命名index
Data.columns = Data.iloc[0]
Data = Data.iloc[1:-1]
Data.index = list(range(len(Data)))
Data[['Charge','Intensity','Score','Retention time']] = Data[['Charge','Intensity','Score','Retention time']].astype(float)
## 展示数据结构与内容
Data
```

```python
## 只保留每个run下鉴定多肽中响应最高的那个
grouped_data = Data.groupby(['Raw file', 'Modified sequence','Charge'])
# 找到Intensity列的最大值
max_intensity = grouped_data['Intensity'].transform('max') ##找到每个分组中Intensity值最大的数据
# 过滤掉Intensity列不是最大值的样本
del_index = Data.loc[Data['Intensity'] != max_intensity].index #找到与最大值不相等的行的index
Data = Data.drop(index=del_index)  ## 删除
print(Data.shape)
```

```python
## 删除每个多肽在不同run之间RT的异常值（基于3σ准则进行异常值筛选，并保留只有一个RT的结果）
mean_RT = Data.groupby('Modified sequence')['Retention time'].transform('mean')
std_RT = Data.groupby('Modified sequence')['Retention time'].transform('std')
threshold_upper = mean_RT + 3 * std_RT
threshold_lower = mean_RT - 3 * std_RT
Data1 = Data.loc[(Data['Retention time'] <= threshold_upper) & (Data['Retention time'] >= threshold_lower)]
Data2 = Data.loc[std_RT[std_RT.isnull()].index]
Data = pd.concat([Data1, Data2],axis=0)
Data = Data.sort_index()
print(Data.shape)
```

```python
## 计算每个多肽在不同run之间RT的RSD，并删除RSD>10%的数据
mean_RT = Data.groupby('Modified sequence')['Retention time'].transform('mean')
std_RT = Data.groupby('Modified sequence')['Retention time'].transform('std')
RSD_RT = std_RT / mean_RT *100
RSD_index = Data.loc[RSD_RT > 10].index
Data = Data.drop(index = RSD_index)  ## 删除RSD>10%的数据
print(RSD_index)
print(Data.shape)
```

```python
## 计算每个多肽在不同run之间RT的平均值
Data.index = list(range(len(Data)))
mean_RT = Data.groupby('Modified sequence')['Retention time'].transform('mean').round(2)
Data['Retention time'] = mean_RT
## 删除score得分小于40分的结果
Data = Data.query('Score >= 40')
## 删除Intensity小于1e4的结果
Data = Data.query('Intensity >= 1e4')
## 删除RT小于10 min的结果
Data = Data.query("`Retention time` > 10")
print(Data.shape)
```

```python
## 导入'Deep learning the collisional cross sections of the peptide universe from a million experimental values'文章计算的CCS值
## 通过匹配将Data中的CCS值替换成文章计算校正后的CCS值
CCS = pd.read_csv(r"sourcedata\SourceData_Figure_1.csv")
CCS = CCS[['Modified sequence','Charge','CCS']]
## 利用pd.merge函数将与CCS与Data中在'Modified sequence','Charge'处相同的数据进行合并
Data = Data.drop('CCS', axis=1)  ## 删除maxquant的CCS值
Data = pd.merge(Data,CCS,how="inner",on=['Modified sequence','Charge'])    
Data['CCS'] = Data['CCS'].astype('float').round(2) 
Data['m/z'] = Data['m/z'].astype('float').round(4) 
Data['Mass'] = Data['Mass'].astype('float').round(4) 
Data['Score'] = Data['Score'].astype('float').round(2) 
```

```python
## 查看数据组成
RT_grouped = Data.groupby(['Modified sequence'])
CCS_grouped = Data.groupby(['Modified sequence','Charge'])
MSMS_grouped = Data.groupby(['Modified sequence','Charge','Intensity'])
print('The unique number of seq and RT is:',len(RT_grouped))
print('The unique number of CCS is:',len(CCS_grouped))
print('The number of unique intensity and MS/MS is:',len(MSMS_grouped))
```
